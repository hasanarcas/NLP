```python
from transformers import BertForSequenceClassification
from transformers import BertTokenizer
import torch
import numpy as np
```

```python
# load the model
model = BertForSequenceClassification.from_pretrained("./model_save")
tokenizer = BertTokenizer.from_pretrained("./model_save")

model.to("cuda")
```

```python
spam_mail = 'You are lucky you have just won a free trip to Paris!'
# tokenize and predict class of input_sentence
tokenized_sentence = tokenizer.encode(spam_mail)
input_ids = torch.tensor([tokenized_sentence]).cuda()
with torch.no_grad():
    output = model(input_ids)
label_indices = np.argmax(output[0].to('cpu').numpy(), axis=1) 
print(label_indices)
```
