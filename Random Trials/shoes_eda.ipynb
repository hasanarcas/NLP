```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (18,9)
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.style'] = "oblique"
```

```python
df = pd.read_csv("../../datasets/MEN_SHOES.csv")
df.dropna(inplace= True)
df.reset_index(drop=True, inplace=True)
df
```

```python
for i in range(len(df)):
    df.Current_Price[i] = df.Current_Price[i].replace("₹", "")
    df.Current_Price[i] = df.Current_Price[i].replace(",", "")
    df.How_Many_Sold[i] = df.Current_Price[i].replace(",", "")
df
```

```python
# change How_Many_Sold to numerical
df.How_Many_Sold = df.How_Many_Sold.astype(int)
df.Current_Price = df.Current_Price.astype(int)
df.RATING = df.RATING.astype(float)
```

```python
sns.scatterplot(x='How_Many_Sold', y='Current_Price', data=df)
plt.xticks(rotation=90)
plt.xlabel("Shoes Sold",fontsize=14)
plt.ylabel("Price",fontsize=14)
plt.title("Shoes Sold vs Price Scatter",fontsize=24)
plt.show()
```

```python
# group by brand and plot
df.groupby('Brand_Name').mean()['Current_Price'].plot(kind='bar')
plt.xticks(rotation=90)
plt.xlabel("Brand",fontsize=14)
plt.ylabel("Average Price",fontsize=14)
plt.title("Avarage Price for Brand",fontsize=24)
plt.show()
```

```python
df.groupby('Brand_Name').mean()['RATING'].plot(kind='bar')
plt.xticks(rotation=90)
plt.xlabel("Brand",fontsize=14)
plt.ylabel("Average Score",fontsize=14)
plt.title("Average Rating Score for Brand",fontsize=24)
plt.show()
```
