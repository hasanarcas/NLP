```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.rcParams['figure.figsize'] = (18,9)
```

```python
df = pd.read_csv("../../datasets/OnePieceArcs.csv")
df.head()
```

```python
# show the most long Arc
sorted_by_episodes = df.sort_values(by=['TotalEpisodes'], ascending=False)
print("The most long Arc is " + sorted_by_episodes.iloc[0]['Arc'] + " with " + str(sorted_by_episodes.iloc[0]['TotalEpisodes']) + " episodes.")
```

```python
# plot the distribution of the number of episodes
sns.distplot(df['TotalEpisodes'])
plt.show()
```

```python
print("As we can see the average number of episodes is " + str(np.trunc(df['TotalEpisodes'].mean()*10**2)/(10**2)) + ".")
```

```python
# scatter plot of the number of episodes vs the number of chapters
sns.scatterplot(x='TotalEpisodes', y='TotalChapters', data=df)
plt.show()
```

```python
# bar plot of the number of episodes for each Arc
sns.barplot(x='Arc', y='TotalEpisodes', data=sorted_by_episodes)
plt.xticks(rotation=90)
plt.xlabel("Arc Name")
plt.ylabel("Number of Episodes")
plt.title("Arc vs Total Episode",fontsize=24)
plt.show()
```

```python
sorted_by_pages = df.sort_values(by=['TotalPages'], ascending=False)

sns.barplot(x='Arc', y='TotalPages', data=sorted_by_pages)
plt.xticks(rotation=90)
plt.xlabel("Arc Name")
plt.ylabel("Number of Pages")
plt.title("Arc vs Total Pages",fontsize=24)
plt.show()
```
