```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (18,9)
```

```python
df = pd.read_csv("../../datasets/istanbul water consumption.csv")
df.rename(columns={'İlçe': 'District', '2019 (Tüketim-m3)': "2019", '2018 (Tüketim-m3)': "2018", '2017(Tüketim-m3)': "2017", '2016 (Tüketim-m3)': "2016", '2015 (Tüketim-m3)': "2015"}, inplace=True)
df.replace(',', '', regex=True, inplace=True)
df
```

```python
df["2019"] = df["2019"].astype(float)
df["2018"] = df["2018"].astype(float)
df["2017"] = df["2017"].astype(float)   
df["2016"] = df["2016"].astype(float)
df["2015"] = df["2015"].astype(float)
```

# Water consumed in 2019 in Istanbul

```python
# bar plot of how much water is consumed in each district
sns.barplot(x="District", y="2019", data=df.sort_values(by="2019", ascending=False))
plt.xticks(rotation=90)
plt.xlabel("Distric Name")
plt.ylabel("Consumed Water (m3)")
plt.title("Water Consumption in Istanbul")
plt.show()
```
